```python
##Plot total daily SIC
import pandas as pd
from datetime import date, timedelta

import matplotlib.pyplot as plt
import numpy as np

#import matplotlib.dates as mdates

# https://matplotlib.org/stable/


b = pd.read_csv('.txt', delimiter = ",") #Path to daily total ice production
b = b.to_numpy()
SIC = b[:,0]


sdate = date(2016,11,1)   # start date
edate = date(2021,4,1)
day= pd.date_range(sdate,edate-timedelta(days=1),freq='d')

fig = plt.figure(figsize=(13, 7), dpi=80)
ax = plt.axes()
ax.plot(day,SIC)


ax.set_xlim(xmin=day[0],xmax=day[1611])#day[0] is first day of. 213 bc taking 1apr as first day

plt.xlabel('Date')
plt.ylabel('Total Sea Ice Concentration (% of each cell)')

fig.savefig('/.jpg') #Path to save
```
