```python
##CUMULATIVE ICE PRODUCTION (Fig 5)
import pandas as pd
from datetime import date, timedelta

import matplotlib.pyplot as plt
import numpy as np

a = np.loadtxt('.txt',delimiter=',',skiprows=1) #Path to Daily Ice Production data
 #The text file starts 1st March so here subtract what's accumulated through march
y2017 = a[31:,5]-a[31,5]
y2018 = a[31:,6]-a[31,6]
y2019 = a[31:,7]-a[31,7]
y2020 = a[31:,8]-a[31,8]
mean2017_20 = np.nanmean([y2017,y2018,y2019,y2020], axis=0)





sdate = date(2019,4,1)   # start date (arbitrary year since on one x axis)
edate = date(2019,11,1)
day= pd.date_range(sdate,edate-timedelta(days=1),freq='d')

fig = plt.figure(figsize=(13, 7), dpi=80)
ax = plt.axes()
ax.plot(day,y2017,label='2017')#,color='blue')
ax.plot(day,y2018,label='2018')#,color='red')
ax.plot(day,y2019,label='2019')#,color='green')
ax.plot(day,y2020,label='2020')#,color='yellow')

ax.plot(day,mean2017_20,'k--',label='Mean')
point = pd.DataFrame({'x': [day[-3]], 'y': [92]})
ax = point.plot(x='x', y='y', ax=ax, kind='scatter',marker='x',s=50, label='1992-2001 (SSM/I, Tamura et al., 2008)', color='black')
point = pd.DataFrame({'x': [day[-3]], 'y': [123]})
ax = point.plot(x='x', y='y', ax=ax, kind='scatter',marker='s',s=50, label='1992-2013 (SSM/I, Tamura et al., 2016)', color='black')
point = pd.DataFrame({'x': [day[-6]], 'y': [90]})
ax = point.plot(x='x', y='y', ax=ax, kind='scatter',marker='v',s=50, label='2003-10 (AMSR-E, Nihashi et al., 2017)', color='black')
point = pd.DataFrame({'x': [day[-9]], 'y': [90]})
ax = point.plot(x='x', y='y', ax=ax, kind='scatter',marker='*',s=50, label='2013-15 (AMSR2, Nihashi et al., 2017)', color='black')


import matplotlib.dates as mdates

ax.xaxis.set_major_formatter(mdates.DateFormatter("%b-%d"))#Change this  xaciss format
ax.set_ylim(ymin=0)
ax.set_xlim(xmin=day[0],xmax=day[213])#day[0] is first day of

plt.xlabel('Date')
plt.ylabel('Cumulative ice production ($km^3$)')
plt.legend(frameon=False)
fig.savefig('.jpg') #Path to fig 5 save


```
